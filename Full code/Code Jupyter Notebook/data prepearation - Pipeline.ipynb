```python
# fungsi untuk membaca nama file, memuat file, dan menggabungkan dataset
def read_dataset(dataset_dir):
    """
    Function for read name file, load file, and
    merge datsheet
    """
    dataset = pd.DataFrame()

    for i in tqdm(os.listdir(dataset_dir)):
        dataset = pd.concat([pd.read_csv(dataset_dir + i), dataset])
    
    return dataset
```

# 3. Data Defense

```python
def check_data(input_data, params):
    # check data types
    assert input_data.select_dtypes("object").columns.to_list() == params["object_columns"], "an error occurs in object column(s)."
    assert input_data.select_dtypes("int64").columns.to_list() == params["int_columns"], "an error occurs in int32 column(s)."
    
    # check range of data
    assert input_data.Price.between(params["range_price"][0], params["range_price"][1]).sum() == len(input_data), "an error occurs in price range."
    assert input_data.Bedroom.between(params["range_bedroom"][0], params["range_bedroom"][1]).sum() == len(input_data), "an error occurs in bedroom range."
    assert input_data.Bathroom.between(params["range_Bathroom"][0], params["range_Bathroom"][1]).sum() == len(input_data), "an error occurs in Bathroom range."
    assert input_data.Land.between(params["range_land"][0], params["range_land"][1]).sum() == len(input_data), "an error occurs in Land range."
    assert input_data.Garage.between(params["range_Garage"][0], params["range_Garage"][1]).sum() == len(input_data), "an error occurs in Garage range."
    assert input_data.Installment_BCA.between(params["range_BCA"][0], params["range_BCA"][1]).sum() == len(input_data), "an error occurs in Installment_BCA range."
    assert input_data.Installment_BNI.between(params["range_BNI"][0], params["range_BNI"][1]).sum() == len(input_data), "an error occurs in Installment_BNI range."
    assert input_data.Installment_Mandiri.between(params["range_Mandiri"][0], params["range_Mandiri"][1]).sum() == len(input_data), "an error occurs in Installment_Mandiri range."
```

```python
# if data not found error, the mean are suitable with we are design
check_data(data_analisa, params)
```
